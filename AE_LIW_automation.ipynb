```python
from pptx import Presentation
from pptx.util import Pt
from datetime import datetime

# Load the PowerPoint presentation
pptx_file = "../../../../Practice Files/powerpoint_practice_files/AE_LIW_practice.pptx"
presentation = Presentation(pptx_file)
fiscalyear_qtr = "FY2023-Q4"

def insert_text_and_date_into_text_box(slide, text_box):
    # Get the current date
    current_date = datetime.now()

    # Format the date as desired (e.g., "January 2024")
    formatted_date = current_date.strftime("%B %Y")

    # Access the text frame of the text box and set the text
    text_frame = text_box.text_frame
    text_frame.clear()  # Clear existing text

    # Combine all text into a single string
    combined_text = (
        "Contact Center Survey\n"
        f"{fiscalyear_qtr}\n\n"
        f"{formatted_date}"
    )

    # Set the combined text as the content of the text frame
    text_frame.text = combined_text

# Apply font and size to the entire text frame
    text_frame.paragraphs[0].font.size = Pt(28)  # Font size for "Contact Center Survey"
    text_frame.paragraphs[0].font.name = 'Arial'

    text_frame.paragraphs[1].font.size = Pt(28)  # Font size for "FY2023-Q4"
    text_frame.paragraphs[1].font.name = 'Tahoma'
    
    text_frame.paragraphs[2].font.size = Pt(16)  # Font size for the date
    text_frame.paragraphs[2].font.name = 'Univers'

    text_frame.paragraphs[3].font.size = Pt(16)  # Font size for the date
    text_frame.paragraphs[3].font.name = 'Univers'

# Example usage
specified_slide_index = 0  # Replace with the index of the specified slide (0-based index)
specified_text_box_index = 0  # Replace with the index of the specified text box (0-based index)

specified_slide = presentation.slides[specified_slide_index]
specified_text_box = specified_slide.shapes[specified_text_box_index]

insert_text_and_date_into_text_box(specified_slide, specified_text_box)

# Save the modified presentation
presentation.save("modified_AE_LIW.pptx")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Given dataset
data = {
    "Quarter": ["Q4 2022", "Q1 2023", "Q2 2023", "Q3 2023"],
    "Yes": ["15%", "19%", "13%", "19%"],
    "No": ["81%", "81%", "85%", "81%"],
    "Do_not_know": ["4%", "0%", "3%", "0%"]
}

# Convert the dataset to a DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame to a suitable format for Seaborn
df_melted = pd.melt(df, id_vars=["Quarter"], var_name="Response", value_name="Percentage")

# Convert percentages to numeric values
df_melted["Percentage"] = df_melted["Percentage"].str.rstrip('%').astype('float')

# Specify the order of responses
response_order = ["Yes", "No", "Do_not_know"]

# Create a Seaborn grouped bar plot with specified order
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Response", y="Percentage", hue="Quarter", data=df_melted, palette="viridis", order=response_order)
plt.title('Grouped Bar Plot')
plt.xlabel('Response')
plt.ylabel('Percentage')
sns.despine(left=True)

# Show the plot
plt.show()
```

#### Extract data from an Excel chart embedded in a powerpoint slide

```python
from pptx import Presentation
import openpyxl
from io import BytesIO

# Load the PowerPoint presentation
pptx_path = "../../../../Practice Files/powerpoint_practice_files/LIW.pptx"
presentation = Presentation(pptx_path)

# Specify the slide index containing the chart (adjust index as needed)
slide_index = 5

# Get the chart shape from the slide
slide = presentation.slides[slide_index]
chart_shape = None
for shape in slide.shapes:
    if shape.has_chart:
        chart_shape = shape
        break

# Check if a chart shape was found
if chart_shape:
    # Extract the embedded Excel workbook data
    excel_data = chart_shape.chart.chart_data.excel_data

    # Create a BytesIO object to load the Excel workbook into openpyxl
    excel_bytes_io = BytesIO(excel_data)
    
    # Load the Excel workbook using openpyxl
    wb = openpyxl.load_workbook(excel_bytes_io, read_only=True)

    # Access the desired worksheet (you may need to modify this based on your Excel structure)
    ws = wb.active

    # Extract data from the worksheet (example: print the first 10 rows)
    for row in ws.iter_rows(max_row=10):
        for cell in row:
            print(cell.value, end='\t')
        print()
else:
    print("No embedded chart found on the specified slide.")
```
